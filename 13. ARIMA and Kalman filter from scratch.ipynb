### Linear Stochastic Models

* **Stochastic process** is a sequence of random variables $\{x_t\}$, indexed by a time subscript.
* The process is said to be **strictly stationary** if the distribution of $\{x_{\tau+1},\ldots,x_{\tau+n}\}$ doesn't depend on $\tau$ for each $n$.
* If the process is stationary, then:
    * $\E x_t=\mu<\infty$ for all $t$;
    * $\cov\left(x_{\tau+t}, x_{\tau+s}\right)=\E\left(x_{\tau+t}-\mu\right)\left(x_{\tau+s}-\mu\right)=\gamma_{|t-s|}$
* There is a bunch of standard methods for analysis of stationary time series: moving-average process, autoregressive process and their combinations. 

### Moving average
The moving-average process of order $q$ is defined by the equation:

* $\displaystyle{MA(q): y_n=\mu+\sum_{i=0}^qb_i\varepsilon_{n-i}}$, 

where: 
* $\displaystyle{\E\varepsilon_n=0}$,
* $\displaystyle{\E\varepsilon_n^2<\infty}$, 
* $\displaystyle{\E\varepsilon_n\varepsilon_m=0}$, 
* $\displaystyle{\varepsilon\sim N(0,\sigma_n^2)}$, 
* $\sigma_n^2=1$

```python
class MA(object):
    def __init__(self, mu = 0.0, b = None, sigma = 1.0):
        self.mu = mu
        self.b = b
        self.q = b.shape[0]
        self.var = np.sum (b * b)
        self.cov = np.array ([np.sum (b[:self.q - k] * b[k:self.q]) for k in range (self.q)])
        self.sigma = sigma

    def simulate(self, n):
        self.eps = norm.rvs (0, self.sigma, size = n)
        eps = self.eps[::- 1]
        self.h = np.array ([self.mu + np.sum (self.b * eps[i:i + self.q]) for i in range (n - self.q)])
        return self.h

    def fit(self):
        self.sample_average = np.mean (self.h)
        self.sample_var = np.var (self.h)
        n = len(self.h)
        self.sample_cov = np.array([np.cov(self.h[k:], self.h[:n - k])[0][1] for k in range(n - 1)])
```

```python
mu = 10.
b = np.array([1,-2,3,-4,5,-6])
n = 100
ma = MA(mu, b)
plt.plot(ma.simulate(n), label='MA')
plt.plot(ma.eps[::-1], label='noise')
plt.title('Moving-average process')
plt.legend();
```

**Actual moments of the MA process:**
* Mean: $\displaystyle{\E y_n=\mu}$, 
* Variance: $\displaystyle{\D y_n = \sum_{i=0}^q b_i^2}$,
* Autocovariance: $\displaystyle{R(k)=\mathrm{cov}(y_n,y_{n+k})=\sum_{j=0}^{q-k}b_jb_{k+j},k=1,\ldots,q}$

```python
mu = 10.
b = np.array([1, -2, 3, -4, 5, -6])
n = 100
ma = MA (mu, b)
h = ma.simulate (n)
h_aver = np.array ([np.mean (h[:i + 1]) for i in range (len (h) - 1)])
plt.plot(h, alpha=0.3, label='MA')
plt.plot (h_aver, label='mean val est')
plt.title('Moving-average process and mean value estimate')
plt.legend();
```

### Autoregressive model
The autoregressive model of order $p$ is defined by the equation:

$\displaystyle{AR(p): y_n=a_0+\sum_{i=1}^pa_iy_{n-i}+\sigma\varepsilon_n}$

**Actual momemts of AR process**
* Mean value: $\displaystyle{\E y_n=\frac{a_0}{1-\sum_{i=1}^pa_i}}$
* Variance: $\displaystyle{R(0)=\sum_{i=1}^pa_iR(i)+\sigma^2}$
* Autocovariance: $\displaystyle{R(k)=\sum_{i=1}^pa_iR(k-i)}$

**AR in vector and state-space forms**
* Consider autoregressive model:
$$\displaystyle{y_n=a_1y_{n-1}+\ldots+a_py_{n-p}+\varepsilon_{n-1}}$$
* Denote
$\displaystyle{\x_{n-1}=(y_{n-1},\ldots,y_{n-p})^T}$
* Rewrite autoregressive model in vector form:
$$\displaystyle{y_n=(a_1,\ldots,a_p)\x_{n-1}+\varepsilon_{n-1}}$$
* AR model in state-space form:
$$\displaystyle{
\x_n=\left(\begin{array}{ccccc}
a_1 & a_2 & \cdots & a_{p-1} & a_p \\
1 & 0 & \cdots & 0 & 0 \\
\vdots & \vdots & \vdots & \vdots & \vdots \\
0 & 0 & \cdots & 1 & 0
\end{array}\right)\x_{n-1}+
\left(\begin{array}{c}
\varepsilon_{n-1} \\
0 \\
\vdots \\
0
\end{array}\right)
}$$

**Observation equation**

Let's introduce **observation equation**:
$$\displaystyle{y_n=H_n\x_n+\xi_n},$$
where
* $\displaystyle{H_n=(1,0,\ldots,0)}$,
* $\displaystyle{\x_n=(y_n,\ldots,y_{n-p+1})^T}$

#### Inversion method for AR coefficients estimation
Solve the system of equations:
$$\displaystyle{\underbrace{\left(\begin{array}{c}
y_{p+1} \\
y_{p+2} \\
\vdots \\
y_n
\end{array}\right)}_{\y}
=\underbrace{\left(\begin{array}{cccc}
y_p & y_{p-1} & \cdots & y_1 \\
y_{p+1} & y_p & \cdots & y_2 \\
\vdots & \vdots & \vdots & \vdots \\
y_{n-1} & y_{n-2} & \cdots & y_{n-p+1}
\end{array}\right)}_{Y}
\underbrace{\left(\begin{array}{c}
a_1 \\
a_2 \\
\vdots \\
a_p
\end{array}\right)}_{a}
\Rightarrow\hat{a}=(Y^TY)^{-1}Y^T\y}$$

#### Yule-Walker algorithm for AR coefficients estimation:
Solve the system of equations:
$$\displaystyle{\underbrace{\left(\begin{array}{c}
\rho_1 \\
\rho_2 \\
\vdots \\
\rho_p
\end{array}\right)}_{\rho}
=\underbrace{\left(\begin{array}{cccc}
1 & \rho_1 & \cdots & \rho_{p-1} \\
\rho_1 & 1 & \cdots & \rho_{p-2} \\
\vdots & \vdots & \vdots & \vdots \\
\rho_{p-1} & \rho_{p-2} & \cdots & 1
\end{array}\right)}_{R}
\underbrace{\left(\begin{array}{c}
a_1 \\
a_2 \\
\vdots \\
a_p
\end{array}\right)}_{a}
\Rightarrow \hat{a}=R^{-1}\rho},$$
where $\displaystyle{\rho(k)}$ - correlation coefficient, which should be replaced by it's sample variant:
$$\displaystyle{r(k)=\frac{\sum_{i=k+1}^n(y_i-\bar{y}_n)(y_{i-k}-\bar{y}_n)}{\sum_{i=1}^n(y_i-\bar{y}_n)^2}},$$
where $\displaystyle{\bar{y}_n=\frac{1}{n}\sum_{i=1}^ny_i}$ - sample mean

```python
def MAP(X, y, alpha = 0.0):
    m = X.shape[1]
    I = np.identity (m)
    Rxx = np.dot (X.T, X)
    Ryx = np.dot (X.T, y)
    if alpha == 0.0:
        return np.dot (np.linalg.inv (Rxx), Ryx)
    elif alpha > 0.0:
        return np.dot (np.linalg.inv (Rxx + alpha * I), Ryx)
```

```python
def autocorr(X, lag = 1):
    cross_corr = np.correlate (X, X, 'full')
    m = np.floor (len (cross_corr) / 2).astype (int)
    acov = cross_corr[m:m + lag]
    acor = acov / acov[0]
    return acor
```

```python
class AR(object):

    def __init__(self, a = None, sigma = 0.1):
        self.p = a.shape[0]
        self.a = a
        self.H = np.zeros (self.p)
        self.H[0] = 1
        self.set_F(a)
        self.x = np.zeros (self.p)
        self.sigma = sigma

    def set_F(self, a):
        F = np.zeros (self.p * self.p).reshape (self.p, self.p)
        F[:, 0] = a
        for i in range (self.p - 1):
            F[i, i + 1] = 1.
        self.F = F.T

    def yule_walker(self, y):
        n = y.shape[0]
        sample_mean = np.mean (y)
        z = y - sample_mean
        self.r = autocorr (z, lag = self.p + 1)
        A = np.eye (self.p)
        b = self.r[1:]
        for i in range (self.p):
            A[i + 1:, i] = b[:- i - 1]
            if i > 0:
                A[:i, i] = b[:i][::- 1]
        self.fit_a = MAP(A, b)
        return self.fit_a

    def inverse(self, y):
        n = y.shape[0]
        b = y[self.p:]
        m = n - self.p
        if m > 0:
            A = np.zeros (m * self.p).reshape (m, self.p)
            for i in range (self.p):
                A[:, i] = y[self.p - i - 1:n - 1 - i]
            self.fit_a = MAP(A, b)
            return self.fit_a
        else:
            return "Incorrect y dimension"

    def fit(self, y, method = 'inverse'):
        if method == 'inverse':
            return self.inverse (y)
        elif method == 'yule-walker':
            return self.yule_walker (y)

    def simulate(self, n):
        self.eps = norm.rvs (loc = 0, scale = self.sigma, size = n)
        y = np.zeros (n)
        x = self.x
        for i in range (self.p, n):
            x = y[i - self.p:i][::-1]
            x = np.dot (self.F, x)
            y[i] = np.dot (self.H, x) + self.eps[i]
        return y
```

### Kalman filter for dynamic system state vector estimation:
* State-space model consists of 
    * state-transition equation $\x_n=F_n\x_{n-1}+w_{n-1}$, where $\x_n$ is the system state vector at the moment $n$, matrix $F_n$ describes system transitions and system noise $w_n$ has zero mean and covariance matrix $Q_n$;
    * observation equation $\y_n=H_n\x_n+\varepsilon_n$, where $\y_n$ - observations, matrix $H_n$ transforms state vector to observation and measurement noise $\varepsilon_n$ has zero mean and covariance matrix $R_n$.
* State-space model system of equations:
$$\displaystyle{
\left\{\begin{array}{l}
\x_n=F_n\x_{n-1} +w_{n-1}\\
\y_n=H_n\x_n+\varepsilon_n
\end{array}\right.
}$$
* Our goal is to estimate state vector $\x$ based on observations $\y$

**How to construct an estimator?**

**Unbiased estimate**
* An estimate of the state vector at the moment $k$ will combine information from the state vector at the moment $k-1$ and from the measurement at the moment $k$: $$\hat{\x}_{k|k}=L_k\hat{\x}_{k|k-1}+K_k\y_k$$
* Define error vectors:
    * $\tilde{\x}_{k|k}=\hat{\x}_{k|k}-\x_k$
    * $\tilde{\x}_{k|k-1}=\hat{\x}_{k|k-1}-\x_k$
* Rewrite our *model* estimate in terms of error vectors: $$\displaystyle{\tilde{\x}_{k|k}=\left(L_k+K_kH_k-I\right)\x_k+L_k\tilde{\x}_{k|k-1}+K_k\varepsilon_k}$$
* We want an **unbiased** estimate $\E\hat{\x}_{k|k}=\x_k$, which means, that $$L_k=I-K_kH_k$$
* From the last condition follows, that our estimate takes the form $$\displaystyle{\hat{\x}_{k|k}=\left(I-K_kH_k\right)\hat{\x}_{k|k-1}+K_k\y_k}$$
and eventually we get
$$\displaystyle{\hat{\x}_{k|k}=\hat{\x}_{k|k-1}+K_k\left(\y_k-H_k\hat{\x}_{k|k-1}\right)}$$
where $$\displaystyle{\mu_k=\y_k-H_k\hat{\x}_{k|k-1}}$$ is usually called **renewal process**.

**Estimate with the smallest variance**
* Define covariance matrices for the error vectors:
    * $\displaystyle{P_{k|k}=\E\tilde{\x}_{k|k}\tilde{\x}^T_{k|k}}$
    * $\displaystyle{P_{k|k-1}=\E\tilde{\x}_{k|k-1}\tilde{\x}^T_{k|k-1}}$
* Put the expression for the $\hat{\x}_{k|k}$ into the expression for the covariance matrices and obtain
$$\displaystyle{P_{k|k}=\left(I-K_kH_k\right)P_{k|k-1}\left(I-K_kH_k\right)^T+K_kRK_k^T}$$
where we took into account, that 
    * $\E\tilde{\x}_{k|k-1}\varepsilon_k^T=0$
    * $\E\varepsilon_k\tilde{\x}^T_{k|k-1}=0$
* Let's look at our task as at the minimization task for the matrix function
$$\displaystyle{J(K)=\left(I-KH\right)P\left(I-H^TK^T\right)+KRK^T=P-KHP-PH^TK^T+K\left(HPH^T+R\right)K^T}$$
* Solve the equation $\displaystyle{\frac{\partial J}{\partial K}=-PH^T+K\left(HPH^T+R\right)=0}$
* Eventually we get, that $$\displaystyle{K_k=P_{k|k-1}H_k^T\left(H_kP_{k|k-1}H_k^T+R\right)^{-1}}$$

**State prediction**

* State vector prediction is simply a $\hat{\x}_{k+1|k}=F_k\hat{\x}_{k|k}$
* Prediction error is $\tilde{\x}_{k+1|k}=F_k\tilde{\x}_{k|k}+w_k$
* Covariance matrix for the prediction error $P_{k+1|k}=F_kP_{k|k}F_k^T+Q$

**Kalman filter algorithm**

* Prediction step:
$$\displaystyle{
\left\{\begin{array}{l}
\hat{\x}_{k|k-1}=F_k\hat{\x}_{k-1|k-1} \\
P_{k|k-1}=F_kP_{k-1|k-1}F_k^T +Q_k
\end{array}\right.
}$$
* Update step:
    * Renewal process $\mu_k=\y_k-H_k\hat{\x}_{k|k-1}$
    * Observation covariance $S_k=H_kP_{k|k-1}H_k^T+R_k$
    * Kalman gain matrix $K_k=P_{k|k-1}H_k^TS_k^{-1}$
    * Updated state vector estimate $\hat{\x}_{k|k}=\hat{\x}_{k|k-1}+K_k\y_k$
    * Updated covariance estimate $P_{k|k}=(I-K_kH_k)P_{k|k-1}$

```python
def matrix_inv(A):
    if A.shape == ():
        return np.array (1. / A)
    else:
        return np.linalg.inv (A)
```

```python
class KalmanFilter(object):

    def __init__(self, F = None, H = None, sigma = 0.1):
        self.F = F
        self.m = F.shape[0]
        self.H = H
        if self.H.ndim == 2:
            self.k = self.H.shape[0]
        else:
            self.k = 1
            self.H = self.H.reshape (1, self.H.shape[0])
        self.sigma = sigma
        self.x = np.zeros (self.m)
        self.P = np.eye (self.m) * self.sigma


    def set_init(self, x, P):
        self.x = x
        self.P = P

    def predict(self, x, P):
        x_k_1 = np.dot (self.F, x)
        P_k_1 = np.dot (self.F, np.dot (P, self.F.T))
        return x_k_1, P_k_1

    def update(self, y_k, x, P):
        y_for = np.dot (self.H, x)
        self.y_fit.append (y_for)
        mu_k = y_k - y_for
        self.mu.append (mu_k)
        S_k = np.dot (self.H, np.dot (P, self.H.T)) + np.eye (self.k) * self.sigma
        S = matrix_inv (S_k)
        K_k = np.dot (np.dot (P, self.H.T), S)

        self.x_k = x + np.dot (K_k, mu_k)
        m = self.x_k.shape[0]
        I = np.eye (m)
        self.P_k = np.dot ((I - np.dot (K_k, self.H)), P)
        return self.x_k, self.P_k

    def fit(self, y):
        n = y.shape[0]
        x = self.x
        P = self.P
        self.y_fit =[]
        self.mu =[]
        self.track =[]
        for i in range (n):
            x, P = self.predict (x, P)
            x, P = self.update (y[i], x, P)
            self.track.append (x)
        self.y_fit = np.array (self.y_fit)
        self.mu = np.array (self.mu)
        self.track = np.array (self.track)
        return x, P
```

### Simple dynamic system
Define dynamic system, describing object's rectilinear uniform motion in 2D space:
$\displaystyle{
\left\{\begin{array}{l}
\x_n=F_n\x_{n-1} \\
\y_n=H_n\x_n+\varepsilon_n
\end{array}\right.
}$
with state vector $\displaystyle{\x_n=(x,y,v_x,v_y)}$ of current coordinates and velocity projections,
$\displaystyle{F_n=F=\left(\begin{array}{cccc}
1 & 0 & \Delta t & 0 \\
0 & 1 & 0 & \Delta t \\
0 & 0 & 1 & 0 \\
0 & 0 & 0 & 1
\end{array}\right)}$, where $\Delta t$ is the elementary time interval between two consequent time stamps $n$ and $n+1$,
and measurements vector, which components are noisy measurements of current coordinates of the object:
$\displaystyle{H_n=H=\left(\begin{array}{cc}
1 & 0 & 0 & 0 \\
0 & 1 & 0 & 0
\end{array}\right)}$

```python
class DynamicSystem(object):

    def __init__(self, x = None, F = None, H = None, sigma = 1.0):
        self.x = x
        self.F = F
        self.H = H
        self.sigma = sigma

    def simulate(self, n):
        self.y =[]
        self.track =[]
        x = self.x
        for i in range (n):
            x = np.dot (self.F, x)
            self.y.append (np.dot (self.H, x))
            self.track.append (x)
        return np.array (self.y)
```

```python
F = np.array ([[1, 0, 1, 0],[0, 1, 0, 1],[0, 0, 1, 0],[0, 0, 0, 1]])
H = np.array ([[1, 0, 0, 0],[0, 1, 0, 0]])
x = np.array([0.5, 0.5, 0.2, 0.3])
ds = DynamicSystem (x, F, H)
n = 100
y = ds.simulate(n) + norm.rvs(loc=0, scale=5., size=n * 2).reshape(n, 2)
```

Use Kalman filter for online estimation of the state vector

```python
x = np.zeros (4)
P = np.eye (len (x)) * 1.
kf = KalmanFilter (F, H)
kf.set_init (x, P)
x_k, P_k = kf.fit (y)
```

Plot measurements, estimation path and true trajectory:

### Kalman filter for AR model coefficients estimation
* Consider AR model:
$y_n=a_1y_{n-1}+\ldots+a_py_{n-p}+\varepsilon_n$
* Write it in the state-space form:
$$\displaystyle{
\left\{\begin{array}{l}
\x_n=F_n\x_{n-1} \\
\y_n=H_n\x_n+\varepsilon_n
\end{array}\right.
},$$
where
$\x_k=(a_1,\ldots,a_p)^T$, $F_n=I_n=I$ - identity matrix, $H_n=(y_{n-1},\ldots,y_{n-p})$

```python
class KalmanFilterAR(object):

    def __init__(self, order=1, sigma=1.):
        self.m = order
        self.F = np.eye(self.m)
        self.H = np.zeros(self.m)
        self.sigma = sigma
        self.x = np.zeros (self.m)
        self.P = np.eye(self.m) * self.sigma ** 2

    def set_init(self, x, P):
        self.x = x
        self.P = P

    def predict(self, x, P):
        x_k_1 = np.dot (self.F, x)
        P_k_1 = np.dot (self.F, np.dot (P, self.F.T))
        return x_k_1, P_k_1

    def update(self, y_k, x, P):
        y_for = np.dot (self.H, x)
        self.y_fit.append (y_for)
        mu_k = y_k - y_for
        self.mu.append (mu_k)
        S_k = np.dot(self.H, np.dot(P, self.H.T)) + np.eye(1) * self.sigma ** 2
        S = matrix_inv (S_k)
        K_k = np.dot (np.dot (P, self.H.T), S)

        self.x_k = x + np.dot (K_k, mu_k)
        I = np.eye(self.m)
        self.P_k = np.dot ((I - np.dot (K_k, self.H)), P)
        return self.x_k, self.P_k

    def fit(self, y):
        n = y.shape[0]
        x = self.x
        P = self.P
        self.y_fit =[]
        self.mu =[]
        self.track =[]
        for i in range(self.m, n):
            x, P = self.predict (x, P)
            self.H = y[i - self.m:i][::-1].reshape(1, self.m)
            x, P = self.update (y[i], x, P)
            self.track.append (x)
        self.y_fit = np.array (self.y_fit)
        self.mu = np.array (self.mu)
        self.track = np.array (self.track)
        return x, P
```

Let's simulate AR process and estimate it's coefficients via Yule-Walker method

```python
a = np.array([0.05, 0.9])
n = 100
ar = AR(a, sigma=1.)
y = ar.simulate (n)
plt.plot(y)
plt.title('AR process')
print('Estimated coefficients (Yule-Walker method): ', ar.fit (y, method = 'yule-walker'))
```

### Autoregressive model of Moving Average

ARMA model of order $(p,q)$ is defined by the equation:

$$\displaystyle{ARMA(p,q): y_n=a_1y_{n-1}+\ldots+a_py_{n-p}+b_1\varepsilon_{n-1}+\ldots+b_q\varepsilon_{n-q}+\varepsilon_n}$$

**ARMA model in state-space form: **
$$\displaystyle{\left\{\begin{array}{l}
y_n=(1,b_1,\ldots,b_{r-1})\x_n \\
\x_n=\left(\begin{array}{ccccc}
a_1 & a_2 & \cdots & a_{r-1} & a_r \\
1 & 0 & \cdots & 0 & 0 \\
\vdots & \vdots & \vdots & \vdots & \vdots \\
0 & 0 & \cdots & 1 & 0
\end{array}\right)\x_{n-1}+
\left(\begin{array}{c}
\varepsilon_n \\
0 \\
\vdots \\
0
\end{array}\right)
\end{array}\right.},$$ 
where $r=\max(p,q+1)$

#### Hannan-Rissanen algorithm for ARMA(p,q) coefficients estimation:
Take $r=\max(p,q)$ and set $a_k=0$ for $k > p$ and $b_k=0$ for $k>q$.
1. Estimate AR coefficients via Yule-Walker method, solving the system of equations:
$$\displaystyle{\underbrace{\left(\begin{array}{c}
\rho_1 \\
\rho_2 \\
\vdots \\
\rho_p
\end{array}\right)}_{\rho}
=\underbrace{\left(\begin{array}{cccc}
1 & \rho_1 & \cdots & \rho_{p-1} \\
\rho_1 & 1 & \cdots & \rho_{p-2} \\
\vdots & \vdots & \vdots & \vdots \\
\rho_{p-1} & \rho_{p-2} & \cdots & 1
\end{array}\right)}_{R}
\underbrace{\left(\begin{array}{c}
a_1 \\
a_2 \\
\vdots \\
a_p
\end{array}\right)}_{a}
\Rightarrow \hat{a}=R^{-1}\rho}$$
2. Get vector of estimated errors
$$\displaystyle{\underbrace{\left(\begin{array}{c}
y_{r+1} \\
y_{r+2} \\
\vdots \\
y_n
\end{array}\right)}_{\y}
=\underbrace{\left(\begin{array}{cccc}
y_r & y_{r-1} & \cdots & y_1 \\
y_{r+1} & y_r & \cdots & y_2 \\
\vdots & \vdots & \vdots & \vdots \\
y_{n-1} & y_{n-2} & \cdots & y_{n-r+1}
\end{array}\right)}_{Y}
\underbrace{\left(\begin{array}{c}
\hat{a}_1 \\
\hat{a}_2 \\
\vdots \\
\hat{a}_r
\end{array}\right)}_{\hat{a}}
}$$
Errors vector: $\displaystyle{e=\y-Y\hat{a}}$

3. Solve the system of equations:
$$\displaystyle{\underbrace{\left(\begin{array}{c}
y_{r+1} \\
y_{r+2} \\
\vdots \\
y_n
\end{array}\right)}_{\y}
=\underbrace{\left(\begin{array}{cccccccc}
y_r & \cdots & y_{r-p} & e_r & e_{r-1} & \cdots & e_{r-q} \\
y_{r+1} & \cdots & y_{r-p+1} & e_{r+1} & e_r & \cdots & e_{r-q+1} \\
\vdots & \vdots & \vdots & \vdots & \vdots & \vdots & \vdots & \vdots \\
y_{n-1} & \cdots & y_{n-p+1} & e_{n-1} & e_{n-2} & \cdots & e_{n-q+1}
\end{array}\right)}_{A}
\underbrace{\left(\begin{array}{c}
\\
a \\
\\
b \\
\\
\end{array}\right)}_{\beta}
\Rightarrow \hat{\beta}=(A^{T}A)^{-1}A^{T}\y}$$

```python
class ARMA(object):

    def __init__(self, a = None, b = None, sigma = 1.):
        self.p = a.shape[0]
        self.q = b.shape[0]
        self.r = max (self.p, self.q + 1)
        self.a = np.zeros (self.r)
        self.b = np.zeros (self.r - 1)
        self.a[:self.p] = a
        self.b[:self.q] = b

        self.H = np.zeros (self.r)
        self.H[0] = 1
        self.H[1:] = self.b

        self.G = np.zeros (self.r)
        self.G[0] = 1.

        self.set_F (self.a)
        self.x = np.zeros (self.r)

        self.sigma = sigma

    def set_F(self, a):
        _a = np.zeros (self.r)
        if a.shape[0] == self.p:
            _a[:self.p] = a
        else:
            _a = a
        F = np.zeros (self.r * self.r).reshape (self.r, self.r)
        F[:, 0] = _a
        for i in range (self.r - 1):
            F[i, i + 1] = 1.
        self.F = F.T

    def yule_walker(self, y):
        n = y.shape[0]
        r = self.r + 1
        sample_mean = np.mean (y)
        z = y - sample_mean
        self.acorr = autocorr (z, lag = r + 1)
        self.e = np.zeros (n)
        A = np.eye (r)
        b = self.acorr[1:]
        for i in range (r):
            A[i + 1:, i] = b[:- i - 1]
            if i > 0:
                A[:i, i] = b[:i][::- 1]
        self.fit_a = MAP(A, b)
        m = n - r
        b1 = y[r:]
        A1 = np.zeros (m * r).reshape (m, r)
        for i in range (r):
            A1[:, i] = y[r - i - 1:n - 1 - i]
        self.e[r:] = b1 - np.dot (A1, self.fit_a)
        return self.fit_a

    def hannan_rissanen(self, y):
        n = y.shape[0]
        b = y[self.r + 1:]
        m = n - self.r - 1
        r = self.p + self.q
        if m > 0:
            A = np.zeros (m * r).reshape (m, r)
            for i in range (r):
                if i < self.p:
                    A[:, i] = y[self.r - i:n - i - 1]
                else:
                    A[:, i] = self.e[self.r - i:n - i - 1]
            self.fit_a_b = MAP(A, b)
            self.e_ma = b - np.dot (A, self.fit_a_b)
            self.fit_a = self.fit_a_b[:self.p]
            self.fit_b = self.fit_a_b[self.p:]
            return self.fit_a, self.fit_b
        else:
            return "Incorrect y dimension"

    def fit(self, y):
        self.fit_a_ = self.yule_walker (y)
        fit_a, fit_b = self.hannan_rissanen (y)
        return fit_a, fit_b

    def simulate(self, n):
        eps = norm.rvs (0, self.sigma, size = n)
        y = np.zeros (n)
        x = self.x
        for i in range (self.r, n):
            x = y[i - self.r:i][::-1]
            x = np.dot (self.F, x) + np.dot (self.G, eps[i])
            y[i] = np.dot (self.H, x)
        return y
```

Consider ARMA process

```python
a = np.array([0.1, 0.6])
b = np.array([0.1])
print('ARMA coefficients: \n alphas: {} \n betas: {}'.format(a,b))
n = 100
arma = ARMA (a, b)
y = arma.simulate (n)
plt.plot (y)
plt.title('ARMA process');
```

Estimate ARMA coefficients via Hannan-Rissanen algorithm

```python
coef = arma.fit(y)
print('ARMA coefficient estimates (Hannan-Rissanen):\n alphas: {} \n betas: {}'.format(coef[0], coef[1]))
```

Estimate ARMA coefficients via **Kalman filter** in the form used for estimation of AR coefficients 

```python
m = len(arma.fit_a)
x = np.zeros (m)
P = np.eye (m) * 1.
kf = KalmanFilterAR (order = m)
kf.set_init (x, P)
x_k, P_k = kf.fit (y[m:])
```

```python
print('AR coefficients KF estimates: \n state vector: \n {} \n covariance: \n {}'.format(x_k, P_k))
```

Actual and estimated ARMA process

```python
plt.plot(kf.y_fit, alpha=0.5, color='red', label='estimate', linestyle='dashed')
plt.plot(y[m:], alpha=0.5, label='ARMA')
plt.title('ARMA process and KF estimates')
plt.legend(loc='lower right');
```

### ARIMA - autoregressive integrated moving average model
**I** in ARIMA stands for "integrated", that means, that $\displaystyle{\Delta^r \mathrm{ARIMA}(p,r,q)=\mathrm{ARMA}(p,q)}$

```python
class ARIMA(object):

    def __init__(self, a = None, b = None, r = 0, sigma = 1.):
        self.ARMA = ARMA (a, b, sigma)
        self.r = r

    def simulate(self, n):
        y = self.ARMA.simulate (n)
        for i in range (self.r):
            y = np.cumsum (y)
        return y

    def fit(self, y, r = None):
        if r == None:
            r = self.r
        y_diff = np.diff (y, n = r)
        self.fit_a, self.fit_b = self.ARMA.fit (y_diff)
        return self.fit_a, self.fit_b
```

Simulate ARIMA process:

```python
a = np.array ([.3, - .6])
b = np.array ([0.1])
print('ARIMA coefficients: \n alphas: {} \n betas: {}'.format(a,b))
r = 1
n = 100
arima = ARIMA (a, b, r)
y = arima.simulate (n)
plt.plot(y)
plt.title('ARIMA process');
```

Use standard technique to estimate ARMA coefficients after taking the $r$-th difference of the initial series:

```python
coef = arima.fit(y)
print('ARIMA coefficient estimates:\n alphas: {} \n betas: {}'.format(coef[0], coef[1]))
```

### Autocovariance and spectrum
* Autocovariance: $\displaystyle{\mathrm{cov}(y_n,y_m)=\E(y_n-\E y)(y_m-\E y)}$
* Correlation: $\displaystyle{\mathrm{corr}(y_n,y_m)=\frac{\mathrm{cov}(y_n,y_m)}{\sqrt{\D y_n\D y_m}}}$
* In stationary case there is no dependence on $n$ in $\mathrm{cov}(y_n,y_{n+k})$:

$$\displaystyle{\rho(k)=\mathrm{corr}(y_n,y_{n+k})=\frac{R(k)}{R(0)}}$$

```python
def autocorr(X, lag = 1):
    cross_corr = np.correlate (X, X, 'full')
    m = np.floor (len (cross_corr) / 2).astype (int)
    acov = cross_corr[m:m + lag]
    acor = acov / acov[0]
    return acor
```

Sample mean:
$\displaystyle{\bar{y}=\frac{1}{n}\sum_{k=1}^ny_k}$

Sample variance:
$\displaystyle{s^2=\frac{1}{n-1}\sum_{k=1}^n(y_k-\bar{y})^2}$

Sample correlation coefficient:
$\displaystyle{r(k)=\frac{1}{s^2}\sum_{i=1}^{n-k}(y_i-\bar{y})(y_{i+k}-\bar{y})}$

Sample spectrum function:
$\displaystyle{f_n(\omega)=\frac{s^2}{\pi}\left(1+2\sum_{k=1}^{n-1}r(k)\cos \omega k\right)}$, $0\leq \omega \leq \pi$

```python
def spectral_function(r, s, w):
    f = s / np.pi
    for k in range (1, len (r)):
        f += 2 * r[k] * s / np.pi * np.cos (w * k)
    return f
```

```python
def spectrum(y, n):
    W = np.linspace (0, np.pi, n)
    z = y - np.mean (y)
    r = autocorr (z, lag = len (y))
    s = np.var (y)
    return np.array ([spectral_function (r, s, w) for w in W]), W
```

```python
def acf(y):
    z = y - np.mean (y)
    return autocorr (z, lag = len (y))
```
